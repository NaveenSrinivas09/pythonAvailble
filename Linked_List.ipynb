```python
# Double Linked List

class Node:
    def __init__(self,data):
      self.data=data
      self.prev=None
      self.next=None

class DLL:
    def __init__(self):
      self.head=None

    def insert_begin(self,data):
      nb=Node(data)
      nb.next=self.head
      if self.head is not None:
        self.head.prev=nb
      self.head=nb

    def insert_end(self,data):
      ne=Node(data)
      if self.head is None:
          self.head = ne
          return
      temp=self.head
      while(temp.next):
        temp=temp.next
      temp.next=ne
      ne.prev=temp

    def insert_pos(self,pos,data):
      np=Node(data)
      if pos == 1:
          self.insert_begin(data)
          return
      temp = self.head
      for i in range(1, pos-1):
          if temp is None:
              print("Position out of bounds")
              return
          temp = temp.next
      if temp is None:
          print("Position out of bounds")
          return
      np.next = temp.next
      np.prev = temp
      if temp.next is not None:
          temp.next.prev = np
      temp.next = np


    def delete_begin(self):
      if self.head is None:
          print("Linked List is empty. Cannot delete.")
          return
      temp=self.head
      self.head=temp.next
      if self.head is not None:
        self.head.prev=None
      temp.next=None


    def delete_end(self):
      if self.head is None:
          print("Linked List is empty. Cannot delete.")
          return
      if self.head.next is None:
          self.head = None
          return
      temp=self.head
      while(temp.next is not None):
        temp=temp.next
      temp.prev.next = None
      temp.prev = None

    def length(self):
      temp = self.head
      length = 0
      while temp:
          length += 1
          temp = temp.next
      return length

    def delete_pos(self,pos):
      if self.head is None:
          print("Linked List is empty. Cannot delete.")
          return
      if pos == 1:
          self.delete_begin()
          return
      temp = self.head
      for i in range(1, pos):
          if temp is None:
              print("Position out of bounds")
              return
          temp = temp.next

      if temp is None:
          print("Position out of bounds")
          return

      if temp.prev is not None:
          temp.prev.next = temp.next
      if temp.next is not None:
          temp.next.prev = temp.prev
      temp.next = None
      temp.prev = None


    def display(self):
      if self.head is None:
        print("Linked List is Empty")
      else:
        temp=self.head
        while(temp):
          print(temp.data,"-->",end=" ")
          temp=temp.next
        print("None") # Added None to indicate the end of the list

    def search(self,data):
      temp=self.head
      while(temp):
        if temp.data==data:
          return True
        temp=temp.next
      return False

    def reverse(self):
      temp=self.head
      prev=None
      while(temp):
        next=temp.next
        temp.next=prev
        temp.prev=next
        prev=temp
        temp=next

      self.head=prev
      return self.head

    def reverse_recursive(self, node):
      if node is None or node.next is None:
          self.head = node
          return
      self.reverse_recursive(node.next)
      node.next.next = node
      node.next = None

    def reverse_iterative(self):
      prev = None
      current = self.head
      while current is not None:
          next_node = current.next
          current.next = prev
          current.prev = next_node
          prev = current
          current = next_node
      self.head = prev
      return self.head

    # def length(self):
    #   temp = self.head
    #   length = 0
    #   while temp:
    #       length += 1
    #       temp = temp.next
    #   return length

L=DLL()

L.insert_begin(10)
L.insert_begin(20)
L.insert_end(30)
L.insert_end(40)
L.insert_pos(2,50)

# L.delete_begin()
# L.delete_end()
# L.delete_pos(2) # Changed position to 2 as after deletions, the list is shorter

L.display()
print()
print(L.search(50))

L.length()

L.reverse_iterative()
L.display()

L.reverse_recursive(L.head)
L.display()

L.reverse()
L.display()
```

```python
# Circular Linked List

class Node:
    def __init__(self, data):
        self.data = data
        self.next = None

class CLL:
    def __init__(self):
        self.head = None

    def insert_begin(self, data):
        nb = Node(data)
        if self.head is None:
            self.head = nb
            self.head.next = self.head  # Point to itself for a single node
        else:
            temp = self.head
            while temp.next != self.head:
                temp = temp.next
            temp.next = nb
            nb.next = self.head
            self.head = nb

    def insert_end(self, data):
        ne = Node(data)
        if self.head is None:
            self.head = ne
            self.head.next = self.head
        else:
            temp = self.head
            while temp.next != self.head:
                temp = temp.next
            temp.next = ne
            ne.next = self.head

    def delete_begin(self):
        if self.head is None:
            print("Linked List is empty. Cannot delete.")
            return
        if self.head.next == self.head: # Single node in the list
            self.head = None
            return
        temp = self.head
        while temp.next != self.head:
            temp = temp.next
        temp.next = self.head.next
        self.head = self.head.next

    def delete_end(self):
        if self.head is None:
            print("Linked List is empty. Cannot delete.")
            return
        if self.head.next == self.head: # Single node in the list
            self.head = None
            return
        temp = self.head
        prev = None
        while temp.next != self.head:
            prev = temp
            temp = temp.next
        prev.next = self.head

    def display(self):
        if self.head is None:
            print("Circular Linked List is Empty")
        else:
            temp = self.head
            while True:
                print(temp.data, "-->", end=" ")
                temp = temp.next
                if temp == self.head:
                    break
            print("...") # Indicate it's circular

    def search(self, data):
        if self.head is None:
            return False
        temp = self.head
        while True:
            if temp.data == data:
                return True
            temp = temp.next
            if temp == self.head:
                break
        return False

    def length(self):
        if self.head is None:
            return 0
        temp = self.head
        length = 0
        while True:
            length += 1
            temp = temp.next
            if temp == self.head:
                break

    def delete_pos(self, pos):
      temp = self.head
      prev = None
      for i in range(pos):
          if temp is None:
              print("Position out of bounds")
              return
          prev = temp
          temp = temp.next

    def insert_pos(self, pos, data):
      np = Node(data)
      if pos == 1:
          self.insert_begin(data)
          return
      temp = self.head
      for i in range(1, pos-1):
          if temp is None:
              print("Position out of bounds")
              return
          temp = temp.next
      if temp is None:
          print("Position out of bounds")
          return
      np.next = temp.next
      temp.next = np

    def reverse(self):
      prev = None
      current = self.head
      while current is not None:
          next_node = current.next
          current.next = prev
          prev = current
          current = next_node
      self.head = prev
      return self.head

    def reverse_recursive(self, node):
      if node is None or node.next is None:
          self.head = node
          return
      self.reverse_recursive(node.next)
      node.next.next = node
      node.next = None
      return self.head

    def reverse_iterative(self):
      prev = None
      current = self.head
      while current is not None:
          next_node = current.next
          current.next = prev
          prev = current
          current = next_node

      self.head = prev
      return self.head


# Example Usage
cll = CLL()
cll.insert_begin(10)
cll.insert_begin(20)
cll.insert_end(30)
cll.insert_end(40)

cll.display()
print()
print(cll.search(30))

# cll.delete_begin()
# cll.delete_end()
# cll.display()

cll.insert_pos(2,25)
cll.display()
cll.reverse_iterative()
cll.display()

cll.reverse()
cll.reverse_iterative()
cll.display()
```
